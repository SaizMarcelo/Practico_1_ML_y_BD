```python
from datetime import date, datetime
import re # Biblioteca que permite comparar un formato con un string

# Creamos una funcion  que me controla el formato fecha si es correcto devuelve datetime, de lo contrario None.
def procesar_columna_fechas(fecha):
  #for posicion in df_acumulado['Fecha de Pedido']:
  formato_esperado = r'\d{2}/\d{2}/\d{2} \d{2}:\d{2}'
  try:
    if (re.match(formato_esperado, fecha)):
      return (pd.to_datetime(fecha, format='%m/%d/%y %H:%M'))
    else:
      return None
  # Existen excepciones que generan errores, ante esa situacion se conciedera registro no util.
  except:
    return None

# Aplicamos la funcion a Fecha de Pedido y generamos una nueva columna con los resultados.
df_acumulado['DatosFechaFormateada'] = df_acumulado['Fecha de Pedido'].apply(procesar_columna_fechas)

# Eliminamos los NaT de la columna DatosFechaFormateada
df_acumulado = df_acumulado.dropna(subset=['DatosFechaFormateada'])

# Eliminamos la Columna Fecha de Pedido que ya no es funcional por que no tiene formato adecuado.
df_acumulado = df_acumulado.drop('Fecha de Pedido', axis=1)
df_acumulado = df_acumulado.reset_index()
df_acumulado
```

```python
# Creamos un DataFrame con detalles mes, maximos, minimos, cantidad_total, ingreso_total
data = {
    'Mes':[],
    'Pedidos_Max':[],
    'Pedidos_Min':[],
    'Pedidos_Medio':[],
    'Pedidos_Totales':[],
    'Ingresos_Totales': [],
    'Porcentaje_del_Presupuesto': []
}

dataFrame_1 = pd.DataFrame(data)
total_ventas = df_acumulado['Cantidad Pedida'].sum()

for posicion in range(1,13):
  ventas_mes = df_acumulado.loc[(df_acumulado['Mes']==posicion) & (df_acumulado['Anio']==2019 )]
  filtro_1 = ventas_mes.groupby('Dia')['Cantidad Pedida'].sum()



  df_mes = pd.DataFrame(filtro_1).reset_index()
  pedidos_max = df_mes['Cantidad Pedida'].max()
  pedidos_min = df_mes['Cantidad Pedida'].min()
  pedidos_medio = df_mes['Cantidad Pedida'].mean()

  # Obtenemos la Cantidad total por mes
  ventas_segun_meses = ventas_mes.groupby('Mes')['Cantidad Pedida'].sum()
  df = pd.DataFrame(ventas_segun_meses).reset_index()
  pedidos_totales = df.loc[(df['Mes']==posicion)]['Cantidad Pedida'][0]

  # Obtenemos la Ingresos total por mes
  ingresos_segun_meses = ventas_mes.groupby('Mes')['Precio Unitario'].sum()
  df = pd.DataFrame(ingresos_segun_meses).reset_index()
  ingresos_totales = df.loc[(df['Mes']==posicion)]['Precio Unitario'][0]

  # Obtenemos Procentaje
  porcentaje = (ventas_segun_meses / total_ventas)*100


  nuevo_registro = {
    'Mes': [int(posicion)],
    'Pedidos_Max': [int(pedidos_max)],
    'Pedidos_Min': [int(pedidos_min)],
    'Pedidos_Medio':[float(pedidos_medio)],
    'Pedidos_Totales': [int(pedidos_totales)],
    'Ingresos_Totales': [float(ingresos_totales)],
    'Porcentaje_del_Presupuesto': [float(porcentaje)]
  }

  nuevo = pd.DataFrame(nuevo_registro)
  dataFrame_1 = pd.concat([dataFrame_1, nuevo], ignore_index=True)

dataFrame_1
df_ordenado = dataFrame_1.sort_values(by='Pedidos_Totales', ascending=False)
df_ordenado
```

- ¿Se ha observado algún aumento o disminución significativa en las ventas en días

Para analizar los dias festivos se tuvieron en cuenta la semana anterior a cada dia festivo y la semana en que se celebra el mismo. Excepto por año nuevo que tenemos datos de 9 dias en lugar de 14.

```python
# Clase para facilitar el realizar los graficos
class FechaFestiva:

  def __init__(self, fecha, fecha_inicio, fecha_fin, nombre):
    self.fecha = fecha
    self.fecha_inicio = fecha_inicio
    self.fecha_fin = fecha_fin
    self.nombre = nombre
```

```python
def procesar_fechas_festivas(festivo, df):

    # Cargamos las fechas
  inicio =  pd.to_datetime(festivo.fecha_inicio, format = "%d-%m-%y")
  final = pd.to_datetime(festivo.fecha_fin, format = "%d-%m-%y")
  objetivo = pd.to_datetime(festivo.fecha, format = "%d-%m-%y")
  nombre = festivo.nombre

  # Ubicamos el periodo festivo en el dataframe
  periodo_festivo = df.loc[(pd.to_datetime(df['Fecha']) >= inicio) & (pd.to_datetime(df['Fecha']) <= final)]
  serie_ventas_festivas = periodo_festivo.groupby(['Fecha', 'Nombre Mes', 'Nombre Dia'])['Cantidad Pedida'].sum()
  df_ventas_festivas =  pd.DataFrame(serie_ventas_festivas).reset_index().sort_values('Fecha')
  df_ventas_festivas['Fecha Compuesta'] = df_ventas_festivas['Nombre Dia'] + ' ' + pd.to_datetime(df_ventas_festivas['Fecha']).dt.day.astype(str) + ' de ' + df_ventas_festivas['Nombre Mes']

  # Ubicamos la fecha concreta
  fecha_festiva = df.loc[pd.to_datetime(df['Fecha']) == objetivo]
  serie_fecha_festiva = fecha_festiva.groupby('Fecha')['Cantidad Pedida'].sum()
  df_fecha_festiva = pd.DataFrame(serie_fecha_festiva).reset_index()


  # Graficamos
  figura_ventas_festivas = go.Figure()

  figura_ventas_festivas.add_trace(go.Scatter(x = df_ventas_festivas['Fecha'], #[etiquetas_personalizadas[num] for num in fechas_numericas]
                                              y = df_ventas_festivas['Cantidad Pedida'],
                                              name = 'Perdiodo festivo'))

  figura_ventas_festivas.add_trace(go.Scatter(x=df_fecha_festiva['Fecha'],
                                              y=df_fecha_festiva['Cantidad Pedida'],
                                              marker=dict(size=20, color="MediumPurple"),
                                              name = nombre))


  figura_ventas_festivas.update_layout(
      width=1000,
      height=500,
      title = nombre,
      xaxis = dict(title = "Dias",
                   tickvals = df_ventas_festivas['Fecha'],
                   ticktext = df_ventas_festivas['Fecha Compuesta']),
      yaxis = dict(title = "Total de ventas")
  )
  figura_ventas_festivas.show()

```
